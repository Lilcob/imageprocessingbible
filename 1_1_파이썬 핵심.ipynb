### 연산자(Operator)

```python
# 산술 연산자

a = 10
b = 3
c = a + b    # 더하기, c에 계산 결과 13 저장
d = a - b    # 빼기, c에 계산 결과 7 저장
e = a * b    # 곱하기, e에 계산 결과 30 저장
f = a / b    # 나누기, f에 계산 결과 3.333..저장
g = a % b    # 나누기 후 나머지 구하기, g에 계산 결과 1 저장
h = a // b   # 정수 나누기, h에 계산 결과 3 저장
i = a ** b   # 지수화, i에 1000 저장
```

두 피연산자의 값을 비교하여 부울 결과를 반환하는 비교 연산자가 있습니다. 파이썬은 다음과 같은 비교 연산자를 지원합니다

 같음(==), 같지 않음(!=), 작음(<), 큼(>), 작거나 같음(<=)

 크거나 같음(>=). 예를 들어, 정수 'a = 10'과 'b = 3'이 주어지면 다음과 같이 다양한 비교를 수행할 수 있습니다


```python
# 비교 연산자

j = a == b  # 같음
k = a != b  # 같지 않음
l = a < b   # 보다 작음
m = a > b   # 보다 큼
n = a <= b  # 다음보다 작거나 같음
o = a >= b  # 다음보다 크거나 같음
```

```python
# 논리 연산자

p = True
q = False
r = p and q  # logical AND r에는 False가 저장
s = p or q   # logical OR s에는 True가 저장
t = not p    # logical NOT not에는 False가 저장
```

```python
# 할당 연산자

a += 3   # a = a + 3과 같음
a -= 3   # a = a - 3과 같음
a *= 3   # a = a * 3과 같음
a /= 3   # a = a / 3과 같음
a %= 3   # a = a % 3과 같음
a //= 3  # a = a // 3과 같음
a **= 3  # a = a ** 3과 같음
```

### 제어문

##### n의 계승을 계산하는 프로그램

```python
# for 문

n = 5
factorial = 1
for i in range(1, n + 1):
    factorial *= i
print(factorial)
```

##### n까지 피보나치 수열을 생성하는 코드

```python
# while 문

n = 10
a, b = 0, 1
while a < n:
    print(a, end=' ')
    a, b = b, a + b
```

### 예외 처리

'try' 문을 사용하면 코드가 실행되는 동안 오류를 테스트할 코드 블록을 정의할 수 있습니다.

'예외' 문을 사용하면 시도 블록에서 발생하는 ‘예외'를 포착하고 처리할 수 있습니다.

 파이썬은 'try' 문 뒤에 오는 코드를 프로그램의 "정상적인" 부분으로 실행합니다.

 'except' 문 뒤에 오는 코드는 앞의 try 절에 있는 예외에 대한 프로그램의 응답입니다.

```python
# try-exception 문

try:
    print(x)
except:
    print('예외가 발생했습니다')
```

파이썬은 오류 발생 여부와 관계없이 실행될 코드 블록을 지정할 수 있는 'finally' 키워드를 제공합니다

```python
# finally 문

try:
    print(x)
except:
    print('예외가 발생했습니다')
finally:
    print('try-except 블록이 완료되었습니다')
```

### 함수

파이썬을 더 깊이 파고들수록 코드의 성능, 유연성 및 구성을 크게 향상시키는 프로그래밍 기능인 함수를 만나게 됩니다.

 파이썬에서 함수는 한 가지 특정 작업을 수행하도록 설계된 코드 블록입니다.

  이 장 에서는 파이썬 함수의 개념을 소개하고, 함수를 정의하고 호출하는 방법을 설명하고, 매개변수와 반환값(return)에 대해 논의하고, 함수를 사용하여 코드를 구성하고 재사용성을 높이고 읽기 쉽게 만드는 방법에 대한 예를 보여드립니다.

```python
# 함수 기본형

def greet():
    print('Hello, World!')

greet()
```

```python
# 반환 값이 존재하는 함수 1

def greet(name):
    print(f'Hello, {name}!')

greet('Alice')
```

```python
# 반환 값이 존재하는 함수 2

def square(number):
    return number ** 2

print(square(5))
```

```python
# 반환 값이 존재하는 함수 3

def greet(name, greeting='안녕하세요'):
    print(f'{greeting}, {name}!')

greet('앨리스')
greet('밥', '좋은 아침')
```

### 클래스
파이썬에서 가장 중요한 부분 중 하나인 클래스 파트입니다.

객체 지향 프로그래밍 언어인 파이썬은 클래스 메커니즘을 통해 객체를 표현할 수 있는 힘을 부여합니다.

클래스는 데이터와 함수를 함께 묶는 방법을 제공하며 이를 ‘캡슐화'라고 표현 합니다.

이 섹션에서는 파이썬 클래스와 관련된 핵심 개념, 클래스 생성, 객체 인스턴스화, 속성 사용, 상속 등을 이해해 보겠습니다.

```python
# 클래스 기본형

class Dog:
    def __init__(self, name, breed, age):
        self.name = name
        self.breed = breed
        self.age = age

    def bark(self):
        print(f"{self.name}가 왈왈 짖습니다")

    def eat(self, food):
        print(f"{self.name}가 {food}을 먹습니다.")

    def sleep(self):
        print(f"{self.name}가 자고 있습니다.")
```

```python
my_dog = Dog("Rex", "German Shepherd", 5)
my_dog.bark()
my_dog.eat("음식")
my_dog.sleep()
```

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.__mileage = 0 # private attribute

    def drive(self, miles):
        if miles > 0:
            self.__mileage += miles
        else:
            print("0보다 커야 합니다!")

    def get_mileage(self):
        return self.__mileage
```

```python
# 상속

class ElectricCar(Car):  # ElectricCar는 Car 클래스를 상속받은 것입니다.
    def __init__(self, make, model, year, battery_size):
        super().__init__(make, model, year) # Car의 init 함수 실행
        self.battery_size = battery_size

    def charge(self):
        print("The car is charging.")
```

```python
# 다형성

class Pet:
    def make_sound(self):
        pass

class Dog(Pet):
    def make_sound(self):
        print("왈왈")

class Cat(Pet):
    def make_sound(self):
        print("야옹")

def pet_sound(pet: Pet):
    pet.make_sound()

dog = Dog()
cat = Cat()

pet_sound(dog)
pet_sound(cat)
```
